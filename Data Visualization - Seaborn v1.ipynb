```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
num = np.random.randn(150)
```

```python
#menampilkan Histogram dan KDE
sns.distplot(num)
```

```python
sns.distplot(num, color='green')
```

```python
label_dist = pd.Series(num, name="Bensin Variabel")
```

```python
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist, vertical=True)
```

```python
sns.distplot(label_dist, vertical=True, color='red')
#menampilkan Histogram dan KDE
```

```python
#Hanya menampilkan KDE dan tdk menampilkan histogram
sns.distplot(label_dist, hist=False)
```

```python
#MENAMPILKAN RUG PLOT
sns.distplot(label_dist, rug=True, hist=False)
```

```python
sns.distplot(label_dist, rug=True, hist=False, color='green')
```
